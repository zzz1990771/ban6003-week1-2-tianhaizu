## 1.4 Lists
A list stores multiple values in order. Python uses zero-based indexing, so the first element is at position `0`.


````python
departments = ["Engineering", "Marketing", "Sales"]

print(departments)
print(departments[0])  # first element
print(departments[1])  # second element
````

Lists can be changed. For example, we can replace the first element.


````python
vector1 = [1, 2, 3]
vector1[0] = 100
print(vector1)
````

A common beginner surprise: multiplying a Python list by 2 repeats the list. It does **not** multiply each element.


````python
vector1 = [100, 2, 3]
print(vector1 * 2)
````

If we really want to multiply each element, one option is a loop. Another option is a list comprehension. This is useful to see, but you do not need to master it today.


````python
# Loop version
new_vector = []
for x in vector1:
    new_vector.append(x * 2)
print(new_vector)

# Optional: list comprehension version
[x * 2 for x in vector1]
````

## 1.5 Dictionaries
A dictionary stores values using key-value pairs. Dictionaries are useful because many datasets also use names or labels to access values.


````python
employee = {
    "name": "Bob",
    "department": "Marketing",
    "salary": 78000
}

print(employee["name"])
print(employee["department"])
print(employee["salary"])
````

## 1.6 A Quick Note on Errors and Debugging
Programming is error-prone. Errors are normal.

Common error types include:
- **Syntax errors:** Python cannot understand the code structure.
- **Runtime errors:** the code starts running but fails during execution.
- **Semantic errors:** the code runs, but the result is not what you intended.

Good debugging habits:
- Read the error message.
- Check the line where the error happened.
- Run code one cell at a time.
- Use `help()` when needed.
- Ask a clear question with the exact error message.


````python
help(print)
````

# Part 2: What is Pandas?

Pandas is one of the most important Python tools for data analysts and data scientists. It helps us work with tabular data: rows and columns.

Pandas is useful because it can:
- load data from CSV, Excel, databases, and other sources;
- store data in DataFrames;
- inspect, filter, summarize, group, clean, and transform data;
- connect later to modeling tools such as Scikit-learn.


## 2.1 Import Pandas
By convention, we import Pandas using the shorter name `pd`.


````python
import pandas as pd
print("Pandas imported successfully!")
````

## 2.2 Series and DataFrame
The two main Pandas objects are:

- **Series:** one column of data.
- **DataFrame:** a table made of multiple Series.

Most of the time in this course, we will work with DataFrames.


````python
sample = pd.DataFrame({
    "employee_id": [1, 2, 3],
    "department": ["Sales", "Marketing", "Engineering"],
    "salary": [60000, 75000, 90000]
})

sample
````

A single column from a DataFrame is a Series.


````python
sample["department"]
````

## 2.3 Methods vs Attributes
In Pandas, you will see both methods and attributes.

- A **method** uses parentheses because it performs an action: `df.head()`.
- An **attribute** does not use parentheses because it stores information: `df.shape`.

This distinction matters because `df.shape` and `df.shape()` are not the same.


````python
print("DataFrame shape attribute:")
print(sample.shape)

print("DataFrame head method:")
sample.head()
````

# Part 5: Data Profiling Checklist

When we profile data, we ask questions like:

1. How many rows and columns are there?
2. What are the data types?
3. Are there missing values?
4. What are the minimum, maximum, average, and typical values for numerical columns?
5. Are categorical values consistent?
6. Are there duplicate IDs or duplicate rows?

We are still diagnosing the data. We are not fixing it yet.


## 5.1 Missing Values
Missing values are often shown as `NaN` in Pandas. They may come from blank cells, system errors, unavailable information, or values that were never collected.


````python
df.isna().sum()
````

It is often helpful to turn missing counts into percentages.


````python
missing_summary = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_percent": df.isna().mean() * 100
})

missing_summary
````

## 5.2 Numerical Summary with `describe()`
`describe()` provides a quick summary of numerical columns.

It includes:
- count of non-missing values;
- mean;
- standard deviation;
- minimum;
- 25th, 50th, and 75th percentiles;
- maximum.


````python
df.describe()
````

## 5.3 Individual Numerical Summary Methods
Sometimes we want one statistic for one column.

The exact column names in your dataset matter. If a column name below does not exist, check `df.columns` and adjust the code.


````python
# Example: summary statistics for age
print("Count:", df["age"].count())
print("Mean:", df["age"].mean())
print("Median:", df["age"].median())
print("Minimum:", df["age"].min())
print("Maximum:", df["age"].max())
print("Standard deviation:", df["age"].std())
````

## 5.4 Categorical Summary with `value_counts()`
For categorical variables, use `value_counts()` to see how often each value appears.


````python
df["department"].value_counts()
````

We can also include missing values in the count by using `dropna=False`.


````python
df["department"].value_counts(dropna=False)
````

## 5.5 Unique Values
`nunique()` counts the number of unique values in a column.


````python
df["department"].nunique()
````

## 5.6 Duplicate Records and Duplicate IDs
A full duplicate row means every value in the row is repeated. A duplicate ID means a column that should uniquely identify records appears more than once.

Both can be important, but they are not always the same issue.


````python
# Count fully duplicated rows
full_duplicate_rows = df.duplicated().sum()
print(f"Number of fully duplicated rows: {full_duplicate_rows}")
````

````python
# Count duplicate employee IDs
employee_id_duplicates = df["employee_id"].duplicated().sum()
print(f"Number of duplicated employee_id values after the first occurrence: {employee_id_duplicates}")
````

`value_counts()` can help us see which IDs appear more than once.


````python
df["employee_id"].value_counts().head(10)
````
